**Praktik PySpark Lanjutan**

```python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
df.show()

df_filtered = df.filter(df['Usia'] > 30)
df_filtered.show()

from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

df_sorted = df.orderBy("Usia", ascending=False)
df_sorted.show()
```

**TUGAS 2 DAN Latihan**

```python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34,"Junior Progammer", "Basket","Laki-laki"),
        ("Budi", 23, "Polisi", "Sepak Bola","Laki-laki"),
        ("Citra", 29,"Dokter Hewan","Membaca buku","Perempuan"),
        ("Dina", 45, "Teknisi Jaringan","Membaca buku","Perempuan")]

columns = ["Nama", "Usia","Pekerjaan","Hobi","Gender"]
df = spark.createDataFrame(data, columns)

print ("Data keseluruhan")
# Menampilkan DataFrame
df.show()

print ('Usia >30')
# Filtering data
df_filtered = df.filter(df['Usia'] > 30)
df_filtered.show()

print ("Menghitung rata-rata")
# Menghitung rata-rata usia
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

print ("Urut berdasarkan usia Terkecil/Terbesar")
# Mengurutkan data berdasarkan usia
df_sorted = df.orderBy("Usia", ascending=True)
df_sorted.show()

print ("Urut berdasarkan usia Terkecil/Terbesar")
# Mengurutkan data berdasarkan usia
df_sorted = df.orderBy("Usia", ascending=False)
df_sorted.show()
```

**Konsep ke 2 untuk latihan dan sebagai tugas.**

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import avg

# Memulai Spark session
spark = SparkSession.builder.appName("AnotherBigDataPractice").getOrCreate()

# Membuat DataFrame lain dengan data berbeda
data = [("Bandung", "Jawa Barat", 2500000),
        ("Surabaya", "Jawa Timur", 2800000),
        ("Yogyakarta", "DIY", 500000),
        ("Kota Magelang", "Jawa Tengah", 128500),
        ("Kota Semarang","Jawa Tengah", 1700000)]
columns= ["Kota", "Provinsi", "Populasi"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame baru
print("Data keseluruhan")
df.show()

# Filtering data (example based on numeric column, adapting to 'Populasi')
print("Populasi > 1000000")
df_filtered = df.filter(df['Populasi'] > 1000000)
df_filtered.show()

# Menghitung rata-rata populasi (adapting from avg("Usia"))
print("Menghitung rata-rata populasi")
df.groupBy().agg(avg("Populasi")).show()

# Mengurutkan data berdasarkan populasi (adapting from orderBy("Usia"))
print("Urut berdasarkan Populasi Terkecil/Terbesar")
df_sorted_asc = df.orderBy("Populasi", ascending=True)
df_sorted_asc.show()

print("Urut berdasarkan Populasi Terbesar/Terkecil")
df_sorted_desc = df.orderBy("Populasi", ascending=False)
df_sorted_desc.show()
```

**TUGAS 4 dan latihan Pandas Lanjutan**

```python
import pandas as pd

# Membuat DataFrame Pandas
data_pandas = {"Nama": ["Ali", "Budi", "Citra", "Dina"], "Usia": [34, 23, 29, 45]}
df_pandas = pd.DataFrame(data_pandas)

# Membuat DataFrame kedua
data_pandas_2 = {"Nama": ["Ali", "Budi", "Citra", "Dina"], "Pekerjaan": ["Dokter", "Guru", "Insinyur", "Perawat"]}
df_pandas_2 = pd.DataFrame(data_pandas_2)

# Join antara dua DataFrame
df_joined = pd.merge(df_pandas, df_pandas_2, on="Nama")
print(df_joined)

# Menghitung statistik deskriptif
print(df_pandas.describe())

# Plotting Data
import matplotlib.pyplot as plt
df_pandas['Usia'].plot(kind='bar')
plt.show()
```

**TUGAS 4**

```python
import pandas as pd

# Membuat DataFrame Pandas
data_1 = {"Kota": ["Bandung", "Surabaya", "Yogyakarta", "Kota Magelang", "Kota Semarang"],
          "Populasi": [2500000, 2800000, 500000, 128500, 1700000]}
df_pandas = pd.DataFrame(data_1)

# Membuat DataFrame kedua
data_2 = {"Kota": ["Bandung", "Surabaya", "Yogyakarta", "Kota Magelang", "Kota Semarang"],
          "Provinsi": ["Jawa Barat", "Jawa Timur", "DIY", "Jawa Tengah", "Jawa Tengah"]}
df_pandas_2 = pd.DataFrame(data_2)

# Join antara dua DataFrame
df_joined = pd.merge(df_pandas, df_pandas_2, on="Kota")
print(df_joined)

# Visualisasi populasi untuk setiap kota
plt.figure(figsize=(10, 6))
plt.bar(df_joined["Kota"], df_joined["Populasi"], color='skyblue')
plt.xlabel("Nama Kota")
plt.ylabel("Jumlah Populasi")
plt.title("Populasi Penduduk di Berbagai Kota")
plt.xticks(rotation=0) # Label diputar agar tidak tumpang tindih
print("\nVisualisasi Data")
plt.show()
```
